### 1. REGRESI

```python
# Melakukan Import Library pandas untuk mengolah data dengan format.csv
# Mengimpor pandas dan memberikan alias 'pd'
import pandas as pd
```

```python
# Menyimpan nilai data file .csv pada Variabel dataUTS
dataUTS = pd.read_csv('R04_corruptions.csv')
```

```python
# menampilkan dataset
dataUTS
```

```python
# Untuk mengecek dan melihat informasi mengenai type data dan kolom yang ada pada dataset kita
dataUTS.info()
```

### KORELASI DATA

```python
# korelasi merupakan tolak ukur kekuatan hubungan linear antara kedua variabel tersebut
dataUTS.corr()
```

### SPLIT DATA

```python
# untuk untuk melatih model regresi atau melakukan analisis lainnya yang memerlukan pemisahan data ke dalam variabel independen dan dependen.
import numpy as np
x_train = dataUTS['Corruption Perception (X)'].values[:, np.newaxis]
y_train = dataUTS['Government Transparency (Y)'].values
```

### MODELLING

```python
# kita dapat menggunakan model ini untuk membuat prediksi pada model data set yang kita punya
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
x_train_test = [[3.6],[4.3],[2.1]]
PredTransparency = model.predict(x_train_test)
print("Prediksi Transparansi : ", PredTransparency)
```

### VISUALISASI DATA

```python
import matplotlib.pyplot as plt
# plot ini membantu visualisasi hubungan antara variabel "Korupsi" dan "Transparansi"
prediksi_transparansi = model.predict(x_train)
dataframe_baru = pd.DataFrame({'Korupsi': dataUTS['Corruption Perception (X)'], 'Transparansi': prediksi_transparansi})
plt.scatter(dataUTS['Corruption Perception (X)'], dataUTS['Government Transparency (Y)'], color="yellow")
plt.plot(dataframe_baru.Korupsi, dataframe_baru.Transparansi, color="blue")
# untuk menambahkan title
plt.xlabel('Corruptions')
plt.ylabel('Transparency')
plt.title('Plot Regresi Data Korupsi')
# untuk menampilkan grafik data
plt.show()
```

### 2. TREE

```python
dataTREE = pd.read_csv('K01_heart_disease.csv')
```

```python
dataTREE
```

```python
dataTREE.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# mengubah nilai dari kolom tipe string/object menjadi numerik
for col in ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]:
    label_encoder = LabelEncoder()
    
    dataTREE[col] = label_encoder.fit_transform(dataTREE[col])
```

### PREPROCESSING DATA

```python
dataTREE.info()
```

### SPLIT DATA

```python
x = dataTREE.drop(["HeartDisease"], axis=1)
y = dataTREE['HeartDisease']
```

```python
# membuat data training dan testing
from sklearn.model_selection import train_test_split
```

```python
# 85:15
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

### Modelling

```python
from sklearn.tree import DecisionTreeClassifier
modeIDT = DecisionTreeClassifier(max_depth=2)
modeIDT.fit(x_train, y_train)
```

```python
hasil_prediksi_dt = modeIDT.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Akurasi Model: ', accuracy_score(y_test, hasil_prediksi_dt) * 100)
```

```python
# visualisasi model decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(40,20))
plot_tree(modeIDT, fontsize=25)
plt.title('Visualisasi Decision Tree', fontsize=50)
plt.show()
```
